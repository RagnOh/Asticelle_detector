```python
import os
import shutil
import random

# Cartelle dei dati
data_dir = "/content/yolov5/DataYolo"  # Sostituisci con il percorso reale
images_dir = os.path.join(data_dir, "images")
labels_dir = os.path.join(data_dir, "labels")

# Percorso per il set di validation
validation_dir = os.path.join(data_dir, "validation")
validation_images_dir = os.path.join(validation_dir, "images")
validation_labels_dir = os.path.join(validation_dir, "labels")

# Creare directory per il set di validation se non esistono
if not os.path.exists(validation_dir):
    os.makedirs(validation_dir)
if not os.path.exists(validation_images_dir):
    os.makedirs(validation_images_dir)
if not os.path.exists(validation_labels_dir):
    os.makedirs(validation_labels_dir)

# Elenchi per le immagini e le label
images = []
labels = []

# Popolamento degli elenchi con le coppie immagine-label
for image_filename in os.listdir(images_dir):
    label_filename = os.path.join(labels_dir, image_filename.split(".")[0] + ".txt")
    if os.path.exists(label_filename):
        images.append(image_filename)
        labels.append(label_filename)

# Numero di immagini per il set di validation
num_validation_images = int(len(images) * 0.2)

# Selezione casuale di immagini e label per il set di validation
validation_indices = random.sample(range(len(images)), num_validation_images)

# Spostamento delle immagini e delle label nel set di validation
for i in validation_indices:
    shutil.move(os.path.join(images_dir, images[i]), validation_images_dir)
    shutil.move(labels[i], validation_labels_dir)

# Stampa del numero di immagini in ciascun set
print(f"Numero di immagini di training: {len(images) - num_validation_images}")
print(f"Numero di immagini di validation: {num_validation_images}")
```

```python
import os

# Percorsi delle cartelle
data_dir = "/content/yolov5/DataYolo"  # Sostituisci con il percorso reale
images_dir = os.path.join(data_dir, "images")
labels_dir = os.path.join(data_dir, "labels")
training_dir = os.path.join(data_dir, "training")  # Cartella di destinazione

# Creare la cartella di training se non esiste
if not os.path.exists(training_dir):
    os.makedirs(training_dir)

# Spostamento delle cartelle images e labels nella cartella di training
shutil.move(images_dir, os.path.join(training_dir, "images"))
shutil.move(labels_dir, os.path.join(training_dir, "labels"))

# Stampa di un messaggio di conferma
print("Cartelle images e labels spostate correttamente nella cartella training.")
```
